```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def generate_realistic_scada_data(start_date="2025-11-01", end_date="2025-11-30"):
    date_range = pd.date_range(start=start_date, end=end_date)
    rows = []

    for current_date in date_range:
        date_int = int(current_date.strftime("%Y%m%d"))

        for block in range(1, 97):  # 96 blocks per day
            # 🎯 SCADA-style load curve (sinusoidal demand shape)
            load_curve = 7000 + 2300 * np.sin((block / 96) * np.pi) + np.random.normal(0, 70)

            mp_demand = round(load_curve, 2)
            cz_demand = round(mp_demand * 0.32 + np.random.normal(0, 50), 2)
            ez_demand = round(mp_demand * 0.28 + np.random.normal(0, 40), 2)
            wz_demand = round(mp_demand * 0.40 + np.random.normal(0, 50), 2)

            row = {
                "date": current_date.strftime("%Y%m%d"),
                "block": block,
                "Raw_Frequency": round(np.random.uniform(49.85, 50.05), 6),
                "MP_Schedule": round(mp_demand * np.random.uniform(0.61, 0.78), 2),
                "MP_Drawal": round(mp_demand * np.random.uniform(0.60, 0.82), 2),
                "MP_Demand": mp_demand,
                "CZ_Total_Schedule": round(cz_demand * 0.70, 2),
                "CZ_Demand": cz_demand,
                "EZ_Total_Schedule": round(ez_demand * 0.73, 2),
                "EZ_Demand": ez_demand,
                "WZ_Total_Schedule": round(wz_demand * 0.72, 2),
                "WZ_Demand": wz_demand,
                "ATPS_Gen": round(np.random.uniform(200, 230), 2),
                "SGTP_Gen": round(np.random.uniform(1230, 1300), 2),
                "SGTS_Gen": round(np.random.uniform(0, 10), 2),
                "SINGHAJI_Gen": round(np.random.uniform(370, 420), 2),
                "Total_Thermal_Gen_Ex_Auxillary": round(np.random.uniform(1700, 2300), 2),
                "Total_Hydel": round(np.random.uniform(700, 900), 2),
                "Thermal_Hydel_Gen": round(np.random.uniform(1900, 2900), 2),
                "UI_Rate": round(np.random.uniform(2.9, 4.2), 6),
                "Total_IPP_Gen": round(np.random.uniform(330, 400), 2),
                "Solar": round(np.random.uniform(10, 90), 2) if 20 < block < 75 else 0,
                "Wind": round(np.random.uniform(15, 120), 2),
                "CZ_Demand_Avg": cz_demand,
                "EZ_Demand_Avg": ez_demand,
                "WZ_Demand_Avg": wz_demand,
                "MP_Demand_Avg": mp_demand,
                "ISP_Hydro": round(np.random.uniform(0.45, 0.7), 6),
                "OSP_Hydro": round(np.random.uniform(250, 350), 2),
                "date_int": date_int,
            }

            rows.append(row)

    df = pd.DataFrame(rows)
    return df

df_november_scada = generate_realistic_scada_data()
df_november_scada.to_csv("SCADA_November_2025_synthetic.csv", index=False)

print("🎉 SCADA-style synthetic dataset generated successfully!")
print(df_november_scada.head(10))
```
